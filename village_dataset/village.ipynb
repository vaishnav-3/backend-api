```python
import pandas as pd
import numpy as np

# Reading the CSV file
lgd_blocks = pd.read_csv("Maharashtra.csv")

# Check the first few rows and the number of columns
print(lgd_blocks.head())          # View the first few rows
print(lgd_blocks.columns)         # Print the column names
print(lgd_blocks.shape)           # Print the shape of the DataFrame

# Rename the columns according to the correct number and names
lgd_blocks.columns = ['s.no', 'state_code', 'state_name', 'district_code', 'district_name', 'block_code', 'block_version', 'block_name', 'col9', 'col10', 'col11', 'col12']

# Selecting specific columns (ensure they exist)
lgd_blocks = lgd_blocks[['state_code', 'state_name', 'district_code', 'district_name', 'block_code', 'block_name']]

# Convert columns to string to avoid TypeError
lgd_blocks['state_name'] = lgd_blocks['state_name'].astype(str)
lgd_blocks['district_name'] = lgd_blocks['district_name'].astype(str)
lgd_blocks['block_name'] = lgd_blocks['block_name'].astype(str)

# Creating 'block_key'
lgd_blocks['block_key'] = lgd_blocks['state_name'] + '-' + lgd_blocks['district_name'] + '-' + lgd_blocks['block_name']

# Lowercasing and cleaning the 'block_key'
lgd_blocks['block_key'] = lgd_blocks['block_key'].str.lower()
lgd_blocks['block_key'] = lgd_blocks['block_key'].str.replace(r"[^a-zA-Z0-9\-]", '', regex=True)

# Setting 'block_key' as index
lgd_blocks = lgd_blocks.set_index('block_key')

# Display the DataFrame
lgd_blocks
```
